```python
from CircularGraph import CircularGraph
g3 = CircularGraph(
    mat_path=Path("Main__Data", "mat_scf100.mat"),
    mat_type="matrix",
    labels="Schaefer 100",
    secondary_labels="Schaefer 100",
    color_palette=Path("Color Palettes","yeo_7_network_colors.csv")
)

print(g3.mat.shape)
print(g3.labels)
print(g3.secondary_labels)
print(g3.color_palette)
```

```python
from thresholds import thresholds as th
thresh = th.PositiveNegativePercentileThreshold(99, True, 99, True)
g3.mask = thresh.apply(g3.mat).edge_mask
```

```python
_ = g3.plot(label=False, sec_label='Bracket', edge_color_method="Uniform", edge_color='#BBBBBB', radius=2, sec_label_size=8, sec_label_font='AppleGothic')
g3.savegraph('test_graph.png', dpi=500)
```
